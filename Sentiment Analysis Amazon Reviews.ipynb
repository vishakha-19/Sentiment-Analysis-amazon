```python
import numpy as np 
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
```

```python
df = pd.read_csv(r'C:\Users\lenovo\Desktop\vishakha\amazon_reviews_1.csv')
df.head()
```

```python
df.shape
```

```python
# Data preparation 
df.dropna(inplace = True)
```

```python
x = df['review']
y = df['review_category']
```

```python
import re
```

```python
# Text cleaning

x_new = []

for review in x:
    review = re.sub('[^a-zA-Z ]', '', review) # punctuation, special characters, number removal 
    review = review.lower()                  # Convert all characters to lowercase
    x_new.append(review)                     # finally adding it to our new list
    
```

```python
'Very bad $Service !!!! you can go to h%%l I was waiting for the deliver for past 23 days '
```

```python
x_new[45] 
```

```python
x_train, x_test, y_train, y_test = train_test_split(x_new, y, test_size = 0.2, random_state=234)
```

```python
# Bag of Words
vector = CountVectorizer(stop_words = 'english').fit(x_train)
```

```python
vector.get_feature_names()
```

```python
len(vector.get_feature_names())
```

```python
x_transform = vector.transform(x_train)
x_transform
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model_lr = LogisticRegression()
model_lr.fit(x_transform, y_train)
```

```python
predictions = model_lr.predict(vector.transform(x_test))
print(classification_report(predictions, y_test))
```

```python
r1 = 'The product was very bad'
predict_r1 = model_lr.predict(vector.transform([r1]))
predict_r1
```

```python
# Tfidf - Term Frequency Inverse document Frequency
vector_tf = TfidfVectorizer(stop_words='english').fit(x_train)
xtf_transform = vector_tf.transform(x_train)
```

```python
model_lr_tf = LogisticRegression()
model_lr_tf.fit(xtf_transform, y_train)
```

```python
prediction_tf = model_lr_tf.predict(vector_tf.transform(x_test))
print(classification_report(prediction_tf, y_test))
```

```python
r2 = 'the product was found in bad condition'
prediction_tf_r2 = model_lr_tf.predict(vector_tf.transform([r2]))
prediction_tf_r2
```
